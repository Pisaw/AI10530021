```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-1.0,1.0,delta)
y = np.arange(-1.0,1.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(X**2 )
Z2 = np.exp(Y**2)
Z=(Z1+Z2)
```

```python
fig,ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title('Simplest default with labets')
```
